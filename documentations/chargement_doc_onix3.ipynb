```python
prefix = "https://hub-dilicom.centprod.com/notices_onix/schema_onix/"
endpoints_list = [
    "ONIX_BookProduct_3.0_reference.xsd",
    "ONIX_BookProduct_3.0_short.xsd",
    "ONIX_BookProduct_CodeLists.xsd",
    "ONIX_XHTML_Subset.xsd",
    "ONIX_XHTML_Subset_reference.xsd",
    "ONIX_XHTML_Subset_short.xsd",
]
```

```python
import requests

for endpoint in endpoints_list:
    url = prefix + endpoint
    print(f"Downloading {url}...")
    response = requests.get(url)
    if response.status_code == 200:
        with open(endpoint, 'wb') as f:
            f.write(response.content)
        print(f"Saved {endpoint}")
    else:
        print(f"Failed to download {endpoint}: HTTP {response.status_code}")
```
